```python
import sklearn.datasets
diabetes = sklearn.datasets.load_diabetes()
print(diabetes.DESCR)
```

```python
x = diabetes.data
print (x[0].size)
```

```python
y = diabetes.target
print (y.size)
```

```python
import matplotlib.pyplot as plt
```

Make a scatter plot using just one of the features in x vs the target variable y. Choosing the first feature:

```python
import matplotlib.pyplot as plt
plt.plot(z,y, "ro")
```

```python
import sklearn.linear_model
import numpy as np
z=[]
for counter in x: 
    z.append(counter[2])
np_array = np.array(z)
```

```python
#Training
import sklearn.linear_model
# Select a linear regression model
model = sklearn.linear_model.LinearRegression()
import sklearn.model_selection

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(
    np_array, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_test = x_test.reshape(-1,1)
x_train = x_train.reshape(-1,1)
```

```python
# Train the model
model.fit(x_train,y_train)
```

```python
# Make a prediction
x_sample = x_test
y_pred = model.predict(x_test)
for i in range(len(y_test)):
    print(y_test[i], y_pred[i])
```

```python
# Visualize what the predictions are for this model

plt.plot(x_test,y_test,'ko')

x_new = np.linspace(-0.1,0.15)
x_new = x_new.reshape(-1,1)
y_pred = model.intercept_ + model.coef_ * x_new
plt.plot(x_new, y_pred)

plt.show()
```

```python
# For this model, we can actually retrieve parameters for our model equation
print(model.coef_)
```

```python
# Now we do have test data!
# and we can calculate our errors on data that were
# not used during training.

from sklearn.metrics import mean_squared_error, r2_score

print('MSE = ', mean_squared_error(y_test, model.predict(x_test)))
print('R^2 = ', r2_score(y_test, model.predict(x_test)))
```
