## Set Up

```python
import sklearn.datasets
import numpy as np

diabetes = sklearn.datasets.load_diabetes()
print(diabetes.DESCR)
x = diabetes.data
y = diabetes.target
```

## K Nearest Neighbours

Use n_neighbors for k-nearest neighbors

```python
import sklearn.neighbors
import sklearn.model_selection
from sklearn.metrics import mean_squared_error

modelkn = sklearn.neighbors.KNeighborsRegressor(n_neighbors=10)
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(
    x, y, test_size=0.2, random_state=42)

modelkn.fit(x_train, y_train)
y_predkn = modelkn.predict(x_test)

print('MSE_knn = ', mean_squared_error(y_test, y_predkn))
```

## Linear Regression

```python
import sklearn.linear_model

modellr = sklearn.linear_model.LinearRegression()
x_trainlr, x_testlr, y_trainlr, y_testlr = sklearn.model_selection.train_test_split(
    x, y, test_size=0.2, random_state=42)

modellr.fit(x_trainlr, y_trainlr)
y_predlr = modellr.predict(x_testlr)

print('MSE_knn = ', mean_squared_error(y_testlr, y_predlr))
```

## Decision Tree Regression

Use max_depth for decision tree

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.tree

from sklearn.metrics import mean_squared_error

model3 = sklearn.tree.DecisionTreeRegressor(max_depth=10)

x_train3, x_test3, y_train3, y_test3 = sklearn.model_selection.train_test_split(
    x, y, test_size=0.2, random_state=42)

model3.fit(x_train3,y_train3)
y_pred3 = model3.predict(x_test3)
print('MSE_tree = ', mean_squared_error(y_test3, y_pred3))
```

## Random Forest Regression

Use max_depth for random forest
